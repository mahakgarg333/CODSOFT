````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

````python
#loading the dataset to a pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
````

````python
#distribution of legit transaction and fradulent transactions
credit_card_data['Class'].value_counts()
````

This dataset is highly unbalanced. We can not feed our model this dataset. So we need to make some of the changes.


0 --> Normal transaction

1--> Fradulent transaction

````python
#separating the data for analysis
#here we have created 2 variables legit and fraud.
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
````

````python
print(legit.shape)
print(fraud.shape)
````

````python
#statistical measures of the data
legit.Amount.describe()
````

````python
fraud.Amount.describe()
````

````python
#compare the values of both transactions
#comparing mean values
credit_card_data.groupby('Class').mean()
````

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and fradulent transactions.

No of fradulent transactions --> 173

````python
legit_sample = legit.sample(n=173)
````

````python
from google.colab import drive
drive.mount('/content/drive')
````

Concatenating two dataframes

````python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
````

````python
new_dataset.head()
````

````python
new_dataset.tail()
````

````python
new_dataset['Class'].value_counts()
````

````python
new_dataset.groupby('Class').mean()
````

Splitting the data into features and targets

````python
X = new_dataset.drop(columns ='Class', axis=1)
Y = new_dataset['Class']
````

````python
print(X)
````

````python
print(Y)
````

Split the data into training data and testing data

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
````

````python
print(X.shape, X_train.shape, X_test.shape)
````

Model training

Logistic regression


````python
model = LogisticRegression()
````

````python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
````

Model Evaluation


Accuracy score


````python
#accuracy on training data
````

````python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
````

````python
print("Accuracy on training data:", training_data_accuracy)
````

````python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
````

````python
print("Accuracy on testing data:", test_data_accuracy)
````
